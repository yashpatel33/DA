```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Titanic = pd.read_csv("titanic.csv")
```

```python
Titanic.info()
```

```python
Titanic
```

```python
Titanic.isnull().sum()
```

```python
df=Titanic.loc[:,["Fare","Survived","Pclass","Sex","Age","Siblings/Spouses Aboard","Parents/Children Aboard"]]
df.head()
```

```python
a=df["Age"].mean()
df["Age"]=df["Age"].fillna(a)
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df["Sex"]=le.fit(df["Sex"]).transform(df["Sex"])
```

```python
df.head()
```

```python
x=df.loc[:,["Fare","Survived","Pclass","Sex","Age","Siblings/Spouses Aboard","Parents/Children Aboard"]]
x.head()
```

```python
x=df.drop("Survived",axis=1)
```

```python
df.head()
```

```python
y=df.loc[:,["Survived"]]
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,)
```

```python
from sklearn import svm
```

```python
clf =  svm.SVC(kernel='linear')
```

```python
clf.fit(x_train,y_train)
```

```python
clf.support_vectors_.shape
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
df_cm=pd.DataFrame(cm.index=["Survived","Unsurvived"])
```
